## WebScraping using Octoparse

### Webscraping can be done using OCTOPARSE without any coding. This app has both premium and free version which helps to scrape data from any websites like Ecommerece, Jobsite, search engines etc., This is the signin page of Octoparse

### Octoparse provide template gallary where we can scrape data with just one click without url of those sites.

### Just by clicling any template we can get data from those sites which we can export to our local system

### Octoparse also provide customised web scrapping from a site just by giving the link of that site

1st step - Creating web page inside OCTOPARSE by pasting the link

#### scrapes the data from multiple pages of the site   

2nd step - Creating pagination (this helps to scrape data from multiple pages)
           and looping all the pages
           
           
Octoparse provides ability to autodetect the whole page and scrape data. we can also choose the required data from the page        

#### Export to local system

we can export the data to our local system as EXCEL, CSV, HTML,JSON files
