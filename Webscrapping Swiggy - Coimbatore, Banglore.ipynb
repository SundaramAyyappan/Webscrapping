```python
from autoscraper import AutoScraper
```

```python
## swiggy coimbatore data
url="https://www.swiggy.com/city/coimbatore"

wanted_list=["MSR Biriyani Hotel", "3.8", "₹200 FOR TWO"]
```

```python
scraper=AutoScraper()
result=scraper.build(url,wanted_list)
print(result)
```
